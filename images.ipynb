```python
import matplotlib.pyplot as plt
import numpy as np
```

TemCNN pooling layers and filter size

```python

filter_sizes = [3, 5, 9, 17, 33]
ind = np.arange(len(filter_sizes))
width = 0.1

markers = {
  'GAP': 'o',
  'MP': 'x',
  'AP': '>',
  'MP+GAP': '+',
  'AP+GAP': '<',
}

polling =  ['GAP', 'MP', 'AP', 'MP+GAP', 'AP+GAP']

means =  np.array([
  [90.34, 89.83, 90.19, 88.93, 90.66],
  [89.72, 87.89, 91.94, 93.1, 91.76],
  [91.62, 92.75, 88.62, 92.27, 93.19],
  [90.04, 91.17, 92.13, 88.67, 90.5],
  [91.79, 91.76, 90.33, 91.35, 90.62],
  ])
std = np.array([
  [0.82, 3.11, 0.19, 4.28, 0.76],
  [1.58, 3.44, 1.5, 0.61, 0.54],
  [0.86, 1.07, 6.46, 1.75, 1.18],
  [1.28, 0.45, 0.62, 2.53, 2.77],
  [1.13, 1.63, 3.75, 1.77, 0.74]
])


fig, ax = plt.subplots()
for i,f in enumerate(polling):
    ax.errorbar(ind - (i-2.5)*width, means[i], yerr=std[i], fmt=markers[f], label=polling[i], )

plt.axhline(y=93.74, linestyle = 'dashed', color='magenta', label='baseline')
ax.set_xticks(ind)
ax.set_xticklabels(filter_sizes)
plt.xlabel('Filter Size')
plt.ylabel('Overall Accuracy (%)')
plt.legend()
fig.tight_layout()
plt.show()
```
